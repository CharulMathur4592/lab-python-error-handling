````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]
````

````python
# Step 1: Define the function for initializing the inventory with error handling

def initialize_inventory(products):
    inventory = {}
    
    for product in products:
        valid_quantity = False
        
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                
                valid_quantity = True
            
            except ValueError as error:
                print(f"Error: {error}")
        
        inventory[product] = quantity
    return inventory
````

````python

def get_valid_orders():
    while True:
        try:
            valid_orders = int(input("Enter the number of customer orders: "))
            
            if valid_orders < 0:
                print("Error: Order Number cannot be negative. Try Again!!")
                continue
            return valid_orders
        except ValueError:
            print("Error: Please enter valid number of orders")
````

````python
def get_valid_product(inventory):
    while True:
        product = input("Enter the product name that a customer wants to order: ").strip().lower()

        if product not in products:
            print("Invalid Product. Choose from: ", products)
            continue

        if inventory.get(product, 0) <= 0:
            print(f"Error: '{product}' is not in stock. Choose another product: ")
            continue

        return product
````

````python
def get_customer_orders(inventory):
    num_orders = get_valid_orders()

    if num_orders == 0:
        return set()

    orders = [get_valid_product(inventory) for _ in range(num_orders)]

    customer_orders = set(orders)
    return customer_orders
````

````python
def calculate_order_statistics(customer_orders, products):
    total_products_ordered = len(customer_orders)

    if len(products) == 0:
        percentage_ordered = 0
    else:
        percentage_ordered = (total_products_ordered / len(products)) * 100

    return total_products_ordered, percentage_ordered
````

````python
def update_inventory(customer_orders, inventory):
    
    updated_inventory = {
        product: quantity - (1 if product in customer_orders else 0)

        for product, quantity in inventory.items()

        if quantity - (1 if product in customer_orders else 0) > 0
    }
    return update_inventory
````

````python
def print_order_statistics(order_statistics):
    
    total_products_ordered, percentage_ordered = order_statistics
    
    print("\nOrder statistics: ")
    print("Total products ordered: ", total_products_ordered)
    print("Percentage of products ordered: ", percentage_ordered)
````

````python
def print_updated_inventory(inventory):
    print("\nUpdated Inventory: ")
    [print(f"{product}: {quantity}") for product, quantity in inventory.items()]
````

````python
def get_valid_price(product):
    
    while True:
        try:
            price = float(input(f"Enter the price of the {product}: "))

            if price < 0:
                print("Error: Price cannot be negative. Try again!!")
                continue
            return price

        except ValueError:
            print("Error: Please enter a valid numerical price: ")
````

````python
def calculate_total_price(customer_orders):
    if not customer_orders:
        return 0.0

    prices = {
        product: get_valid_price(product)
        for product in customer_orders
    }
    total_price = sum(prices.values())
    return total_price
````

````python
def main():
    inventory = initialize_inventory(products)

    customer_orders = get_customer_orders(inventory)

    order_stats = calculate_order_statistics(customer_orders, products)
    print()
    print_order_statistics(order_stats)

    print(inventory.items())

    update_inventory(customer_orders, inventory)
    
    print_updated_inventory(inventory)

    total_price = calculate_total_price(customer_orders)
    print("Total Price: ", total_price)

if __name__ == "__main__":
    main()
````
